```python
#6. Adding HDI Category Column
# • Create a new column HDI Category that classifies each country into one of the four official
# Human Development Index groups. The classification should be based on the HDI value for the
# year 2022. Use the following categories and thresholds defined by the United Nations Development
# Programme (UNDP):

# HDI Category HDI Range (hdi)
# Low < 0.550
# Medium 0.550 – 0.699
# High 0.700 – 0.799
# Very ≥ 0.800
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi <= 0.699:
        return "Medium"
    elif hdi <= 0.799:
        return "High"
    else:
        return "Very High"
```

```python
# After creating this new column:
# • verify that all countries are classified correctly
# Check if any country has missing HDI category
# Add HDI_Category column
# • ensure the updated dataframe includes the new category column.
# • Save the final dataframe as HDI category added.csv and include this file in your final
# submission.
hdi_2022_df = df[df["year"] == 2022].copy()
hdi_2022_df["HDI_Category"] = hdi_2022_df["hdi"].apply(classify_hdi)
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
hdi_2022_df.head()
```

```python
#3. Outlier Detection:
#• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.
#HDI:
hdi_southasia_outliers = hdi_southasia_df.copy()

# ---- HDI outlier detection ----
Q1_hdi = hdi_southasia_outliers["hdi"].quantile(0.25)
Q3_hdi = hdi_southasia_outliers["hdi"].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# checking: is this row an HDI outlier?
hdi_southasia_outliers["is_hdi_outlier"] = (
    (hdi_southasia_outliers["hdi"] < lower_hdi) |
    (hdi_southasia_outliers["hdi"] > upper_hdi)
)

# Show only the HDI outliers
hdi_southasia_outliers[hdi_southasia_outliers["is_hdi_outlier"]][
    ["country", "year", "hdi"]
]
#no outliers detected!
```

```python
#GNI:
# ---- GNI per Capita outlier detection ----
Q1_gni = hdi_southasia_outliers["gross_inc_percap"].quantile(0.25)
Q3_gni = hdi_southasia_outliers["gross_inc_percap"].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni

# Create True/False column: is this row a GNI outlier?
hdi_southasia_outliers["is_gni_outlier"] = (
    (hdi_southasia_outliers["gross_inc_percap"] < lower_gni) |
    (hdi_southasia_outliers["gross_inc_percap"] > upper_gni)
)

# Show only the GNI outliers
hdi_southasia_outliers[hdi_southasia_outliers["is_gni_outlier"]][
    ["country", "year", "gross_inc_percap"]
]
#no outliers detected!
```

```python
# Combined outlier of HDI AND GNI:
hdi_southasia_outliers["is_any_outlier"] = (
    hdi_southasia_outliers["is_hdi_outlier"] |
    hdi_southasia_outliers["is_gni_outlier"]
)
```

```python
plt.figure()

# Normal points (not outliers)
normal_points = hdi_southasia_outliers[~hdi_southasia_outliers["is_any_outlier"]]
plt.scatter(
    normal_points["gross_inc_percap"],
    normal_points["hdi"],
    color="blue",
    label="Normal"
)

# Outlier points
outlier_points = hdi_southasia_outliers[hdi_southasia_outliers["is_any_outlier"]]
plt.scatter(
    outlier_points["gross_inc_percap"],
    outlier_points["hdi"],
    color="red",
    label="Outliers"
)

plt.xlabel("GNI per Capita (gross_inc_percap)")
plt.ylabel("HDI (hdi)")
plt.title("South Asia: GNI per Capita vs HDI with Outliers (1.5 × IQR)")
plt.legend()
plt.show()
#note: there is no outliers both in HDI and GNI in our south asia database!
```

```python
#7. Outlier Detection:
#• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using the 1.5 × IQR rule.
#Detecting outliers for South Asia:
southasian_outliers = hdi_southasia_2020_2022.copy()

Q1_southasian_hdi = southasian_outliers["hdi"].quantile(0.25)
Q3_southasian_hdi = southasian_outliers["hdi"].quantile(0.75)
IQR_southasian_hdi = Q3_southasian_hdi - Q1_southasian_hdi

southasian_hdi_lower = Q1_southasian_hdi - 1.5 * IQR_southasian_hdi
southasian_hdi_upper = Q3_southasian_hdi + 1.5 * IQR_southasian_hdi

southasian_outliers["is_hdi_outlier"] = (
    (southasian_outliers["hdi"] < southasian_hdi_lower) |
    (southasian_outliers["hdi"] > southasian_hdi_upper)
)

print("South Asia - HDI outliers:")
print(southasian_outliers[southasian_outliers["is_hdi_outlier"]][
    ["country", "year", "hdi"]
])

# ---- GNI per Capita (South Asia) ----
Q1_southasian_gni = southasian_outliers["gross_inc_percap"].quantile(0.25)
Q3_southasian_gni = southasian_outliers["gross_inc_percap"].quantile(0.75)
IQR_southasian_gni = Q3_southasian_gni - Q1_southasian_gni

southasian_gni_lower = Q1_southasian_gni - 1.5 * IQR_southasian_gni
southasian_gni_upper = Q3_southasian_gni + 1.5 * IQR_southasian_gni

southasian_outliers["is_gni_outlier"] = (
    (southasian_outliers["gross_inc_percap"] < southasian_gni_lower) |
    (southasian_outliers["gross_inc_percap"] > southasian_gni_upper)
)

print("\nSouth Asia - GNI outliers:")
print(southasian_outliers[southasian_outliers["is_gni_outlier"]][
    ["country", "year", "gross_inc_percap"]
])
```
